```python
from sklearn.datasets import make_classification
import torch
```

```python
# creating a syntetic classification dataset using sklearn
X, y = make_classification(
    n_samples = 10,  # Total number of data points (rows).
    n_features= 2, # Each sample has 2 input features -- (x1, x2)
    n_informative = 2, # Out of all features, 2 features actually matter for classification
    n_redundant= 0, # No duplicate or correlated features
    n_classes = 2, # Binary classification problem.
    random_state= 42 # Fixes the randomness so results are reproducible.
)
```

# Converting the data to PyTorch Tensor

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)
```

```python
X
```

```python
y
```

```python
from torch.utils.data import Dataset , DataLoader
```

```python
class CustomDataset(Dataset):

  # constructor
  def __init__(self, features, labels):
    self.features = features
    self.labels = labels

 # length
  def __len__(self):
    return self.features.shape[0]

# getitem
  def __getitem__(self, index):
    return self.features[index], self.labels[index]
```

```python
dataset = CustomDataset(X,y)
```

```python
len(dataset)
```

```python
dataset[0]
```

# DataLoader

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
for batch_features , batch_labels in dataloader:
  print(batch_features)
  print(batch_labels)

  print("--"*30)
```
