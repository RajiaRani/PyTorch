# Building linear model where i have 5 inputs - 1 neuron - 1 output

```python
import torch
import torch.nn as nn
```

```python
class Model(nn.Module):
  def __init__(self, input_features):
    super().__init__()
    # first linear layer
    self.linear = nn.Linear(input_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    sigmoid = self.sigmoid(out)
    return sigmoid
```

```python
# creating demi datasets
input_features = torch.rand(10, 5)
```

```python
# Create model
model = Model(input_features.shape[1])
```

```python
# Run the model
model(input_features)
```

# I have 5 inputs - 3 neurons(hidden layer) - 1 neuron(output)

```python
class Model(nn.Module):
  def __init__(self, input):
    super().__init__()

    self.linear1 = nn.Linear(input, 3)
    self.relu = nn.ReLU()
    self.hidden_layer = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, input):
    layer1 = self.linear1(input)
    relu = self.relu(layer1)
    layer2 = self.hidden_layer(relu)
    sigmoid = self.sigmoid(layer2)

    return sigmoid

```

```python
input = torch.rand(20, 5)
```

```python
model = Model(input.shape[1])
```

```python
model(input)
```

#Using Sequential Container

```python
class Model(nn.Module):
  def __init__(self, x):

    super().__init__()
    self.networks = nn.Sequential(
        nn.Linear(x, 8),
        nn.ReLU(),
        nn.Linear(8, 5),
        nn.ReLU(),
        nn.Linear(5,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )
def forward(self, input):
    out = self.networks(input)
    return out
```

```python
input = torch.rand(40, 10)
```

```python
model = Model(input.shape[1])
```

```python
model(input)
```
